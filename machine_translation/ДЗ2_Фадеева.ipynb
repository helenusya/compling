<h1>Структура системы машинного перевода</h1>

Система обрабатывает пару русский → французский.

Используется трансферный перевод.

<ul>
    1. Создаём двуязычный лексикон
</ul>
<ul>
    2. Создаём список слов для морфологического анализа
</ul>
<ul>
    3. Прописываем функции предобработки
    <ul>
        3.1. Токенизируем текст
    </ul>
    <ul>
        3.2. Проводим частеречную разметку с помощью POS-теггера
    </ul>
</ul>
<ul>
    4. Прописываем функции замен
    <ul>
        4.1. Создаём правила лексической замены
    </ul>
    <ul>
        4.2. Создаём правила синтаксической замены
    </ul>
    <ul>
        4.3.  Создаём правила грамматической замены
    </ul>
</ul>
<ul>
    5. Генерируем результат в виде функции
    <ul>
        5.1. Склеиваем токены в строку
    </ul>
    <ul>
        5.2. Первое слово пишем с большой буквы
    </ul>
    <ul>
        5.3. В конце предл-ия ставим точку
    </ul>
</ul>
<ul>
    6. Прогоняем все функции
</ul>
<ul>
    7. Выводим на экран перевод
</ul>

<h3>Подробнее о языковых правилах: </h3>

*С точки зрения лексики*: необходимо учесть лексическую неоднозначность. Например, слово “ручка” может переводиться с русского разным образом в зависимости от контекста. Это необходимо прописать в правилах.

*С точки зрения синтаксиса*: если в русском порядок слов максимально свободный, то во французском инверсия не всегда уместна. Кроме того, например, в отличие от русского, во французском прилагательное чаще ставится после существительного (а ряд прилагательных, наоборот, ДО него).

*С точки зрения грамматики*: не все грамматические категории, которые есть в русском, существуют во французском. Например, если мы говорим о деепричастных оборотах в их привычном понимании. Их во французском нет, но есть герундий, форма которого будет выбираться по другим правилам.
Есть, например, и другая сложность: в русском нет времени plus-que-parfait, но при переводе на французский обязательно надо учитывать, если какое-то действие произошло раньше другого в прошлом. Аналогично с согласованием времён в косвенной речи.

Важно учитывать, что русский язык является синтетическим, французский — аналитическим. При переводе и создании правил надо пытаться искать такие переводческие трансформации, которые бы максимально сохранили план содержания даже при вынужденном изменении плана выражения.

В связи с тем, что при переводе с русского надо будет учитывать порой очень сложные синтаксические связи (как и порожденную ими неоднозначность) и придавать конечным французским предложениям более структурированный, строгий (с точки зрения порядка слов) вид, можно было бы использовать парсинг на этапе анализа и предварительной обработки.

<h3>Небольшое дополнение, не касающееся домашней работы:</h3>

В ворклисте RBMT.ipynb с семинара при демонстрации прямого перевода мы “принтовали” промежуточные результаты для интерпретации. В строке “лексические соответствия” при этом оказывалось следующее: “Лексические соответствия: ['heute', 'ist', 'sie', 'glücklich']”. Кажется, что этот результат не совсем верный, ведь на этапе поиска лексических соответствий порядок слов еще не должен был поменяться (это происходит на шаг после, в построении верного порядка слов). Т.е. в соответствии с оригиналом на английском ('today', 'she', 'is', 'happy') тут должно было выйти: 'heute', 'sie', 'ist', 'glücklich'. 

Кажется, проблема заключается в том, как в языке Python работает тип данных list (список). Он является изменяемым: если объявить переменную списка и затем ее присвоить другой переменной или, как в нашем случае, передать ее в функцию local_reordering, то ее значение изменится вместе с новой переменной. Поэтому можно скопировать значение исходного списка в новую переменную (result_tokens), чтобы исходный список не менялся, используя конструкцию индексного обращения к списку: result_tokens = target_tokens[:] - так список _target_tokens останется неизменным.
